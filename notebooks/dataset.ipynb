```python
import pandas as pd

# CSV 파일 불러오기
data_path = "../dataset/raw/spotify_data.csv"
df = pd.read_csv(data_path)

# 크기와 샘플 확인
print(df.shape)
print(df.head(3))
```

```python
# 결측치 확인
print(df.isnull().sum())

# 결측치 제거
df = df.dropna()

# track id 중복 제거
df = df.drop_duplicates(subset=["track_id"])
```

```python
import matplotlib.pyplot as plt

# 장르별 비율
genre_counts = df['genre'].value_counts(normalize=True) * 100

# 시각화
plt.figure(figsize=(8,6))
genre_counts.plot(kind='bar')
plt.title("Music Distribution by Genre (%)")
plt.ylabel("Percentage (%)")
plt.xlabel("Genre")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
